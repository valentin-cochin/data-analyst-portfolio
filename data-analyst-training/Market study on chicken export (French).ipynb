## Synthèse des informations existantes sur le sujet

Même si cela peut paraître trivial, commençons par la définition du poulet pour s'assurer d'éviter toute confusion. Selon le CNRTL, c'est un: 
>Jeune coq ou jeune poule, élevé pour la consommation précoce.

La *FAO* nous révèle l'importance croisssante de cet animal :
>Les poulets (Gallus gallus domesticus) ont été domestiqués à partir du coq doré (Gallus gallus) en Asie il y a environ 5000 ans. Les poulets sont sans aucun doute l'espèce de bétail domestique la plus répandue et la plus abondante et sont présents plus ou moins partout où habite l'être humain. Dans les systèmes d'élevage intensifs, ils ont tendance à être élevés pour leurs œufs (pondeuses) ou la viande (poulets de chair), mais dans les systèmes à petite échelle, ils tendent à avoir un double objectif. Avec des cycles de vie courts et des taux de conversion alimentaire potentiellement élevés, la production de poulet peut être rapidement augmentée pour répondre à la demande croissante d'aliments d'origine animale. Ils constituent donc le sous-secteur de l'élevage dont la croissance est la plus rapide.

Sur son site, la *FAO* nous informe également de sa répartition. Sur la carte ci-dessous, on peut voir le nombre de poulets par kilomètre carré en 2010. On peut y déceler les principales régions productrices de poulet.

Le moyen Orient, l’Afrique du nord et sub-saharienne, l’Europe de l'Est, l’Océanie et la Fédération de Russie, avec une production entre 4 et 1,6 millions de tonnes de protéines, ont chacune une part individuelle plus faible à l'échelle mondiale.

```python
# Manipulation
import numpy as np
import pandas as pd

# Visualisation
import matplotlib.pyplot as plt
import seaborn as sns

# Modélisation
from sklearn import decomposition # Regroupe les algos de décompositions matricielles, utilisé pour l'ACP
from sklearn import preprocessing # Pour centrer et réduire les données
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster # Pour le regroupement hiérarchique
```

```python
# Préparation de la table principale
df_fbs = pd.read_csv('Data\\bilans_alim_2013.csv')

# Visualisation de la table
df_fbs.info()
df_fbs.head(10)
```

```python
#Supression de 'Population' dans l'attribut 'Produits'
df_fbs = df_fbs[df_fbs['Produit'] != 'Population']

# Vérifications
print(df_fbs.isnull().values.any()) # Renvoie True s'il existe encore des NAN
df_fbs.info()
```

```python
# Ajout de l'origine de l'aliment (végétale ou animale)
df_origin = pd.read_csv('Data\\groupe_produits.csv')
df_origin = df_origin[df_origin['Groupe de produits'].isin(['Produits Vegetaux', 'Produits Animaux'])]
df_fbs = pd.merge(left=df_fbs,right=df_origin[["Groupe de produits", "Code Produit"]], how='left', left_on='Code Produit', right_on='Code Produit')

# Ajout des continents
df_continent = pd.read_csv('Data\\groupe_pays.csv')
continents = ["Afrique", "Amériques", "Asie", "Europe", "Océanie"]
df_continent = df_continent[df_continent['Groupe de pays'].isin(continents)]
df_fbs = pd.merge(left=df_fbs,right=df_continent[["Groupe de pays", "Pays"]], how='left', left_on='Pays', right_on='Pays')

# Vérifaction des modifications
df_fbs.info()
df_fbs.head(10)
```

```python
# D'après describe il manque le continent pour certaines lignes
# Génération des pays sans continent
df = df_fbs[df_fbs.isnull().any(axis=1)]
df['Pays'].unique()
```

```python
# Par chance, il n'y en a que deux
# Je les rajoute manuellement
df_fbs.loc[df_fbs['Pays'] == 'Chine', 'Groupe de pays'] = 'Asie'
df_fbs.loc[df_fbs['Pays'] == 'Eswatini', 'Groupe de pays'] = 'Afrique'
df_fbs[df_fbs['Pays'].isin(df['Pays'].unique())]
```

Voici la liste des redondances que j'ai détecté :
1. **[Disponibilité intérieure]** = [Production] + [Importations] - [Exportations] + [Variation de stock]
2. **[Nourriture]** = [Disponibilité intérieure] - [Aliments pour animaux] - [Autres Utilisations] - [Pertes] - [Semences] - [Traitement]

On peut combiner ces deux expressions en remplaçant la *Disponibilité intérieure* comme ceci:
> **[Nourriture]** = [Production] + [Importations] - [Exportations] + [Variation de stock] - [Aliments pour animaux] - [Autres Utilisations] - [Pertes] - [Semences] - [Traitement]

```python
# Reformatage de la table des Bilans Alimentaires
# Création de la dataframe principale
df_all = pd.pivot_table(df_fbs, index=['Code Pays', 'Pays', 'Code Produit', 'Produit', 'Groupe de produits', 'Groupe de pays'], columns='Élément', values='Valeur', aggfunc=sum)
df_all = df_all.reset_index()

# Pour éviter les erreurs de calcul
df_all.fillna(0, inplace=True)

df_all.info()
df_all.head(10)
```

Il faut que j'aggrège les lignes pour avoir :
- Disponibilité alimentaire en protéines par habitant -> Disponibilité de protéines en quantité (g/personne/jour)
- Disponibilité alimentaire en calories par habitant -> Disponibilité alimentaire (Kcal/personne/jour)   
- Proportion de protéines d'origine animale par rapport à la quantité totale de protéines dans la disponibilité alimentaire du pays -> Avant d'aggréger, j'ai besoin de créer deux nouvelles colonnes pour la dernière variable.

Je ne vais pas changer les unités car les données seront centrées et réduites ultérieurement.


```python
# J'appelle ces variables dummy car elles n'ont pas de sens en elles-mêmes
# Ce sont des variables intermédiaires
df_all['dummy_veg'] = df_all["Disponibilité de protéines en quantité (g/personne/jour)"].where(df_all["Groupe de produits"] == 'Produits Vegetaux', 0.0)
df_all['dummy_flesh'] = df_all["Disponibilité de protéines en quantité (g/personne/jour)"].where(df_all["Groupe de produits"] == 'Produits Animaux', 0.0)

df_all.info()
df_all.head()
```

```python
# Nombre de pays répertoriés par la FAO
len(df_all['Pays'].unique())
```

```python
# Nombre d'achats pour chaque de chaque client_id
df_all = df_all.groupby([
        'Code Pays',
        'Pays',
        'Groupe de pays'
]).agg({
        'Aliments pour animaux' : "sum",
        'Autres Utilisations' : "sum",
        "Disponibilité alimentaire (Kcal/personne/jour)" : "sum",
        "Disponibilité alimentaire en quantité (kg/personne/an)" : "sum",
        "Disponibilité de matière grasse en quantité (g/personne/jour)" : "sum",
        "Disponibilité de protéines en quantité (g/personne/jour)" : "sum",
        "Disponibilité intérieure" : "sum",
        "Exportations - Quantité" : "sum",
        "Importations - Quantité" : "sum",
        "Nourriture" : "sum",
        "Pertes" : "sum",
        "Production" : "sum",
        "Semences" : "sum",
        "Traitement" : "sum",
        "Variation de stock" : "sum",
        "dummy_veg" : "sum",
        "dummy_flesh" : "sum"
})

df_all = df_all.reset_index()

df_all.info()
df_all.tail()
```

```python
# Le ratio pop_urbaine/totale permet de regrouper deux colonnes
df_pop['pop_urbaine/totale'] = df_pop['Population urbaine'] / (df_pop['Population rurale'] + df_pop['Population urbaine'])
df_pop.head(10)
```

```python
# Je ne garde que deux années pour calculer la variation
df_pop = df_pop[df_pop['Année'].isin([2012, 2013])]

list_values = ['Population rurale', 'Population totale', 'Population urbaine', 'pop_urbaine/totale']

# Je pivote la table
df_pop = pd.pivot_table(df_pop, index=['Code zone', 'Zone'], columns='Année', values=list_values, aggfunc=sum)

df_pop.head()
```

```python
def flatten_cols(df):
    '''
    Purpose : Flatten a hierarchical index in columns
    Found on : https://stackoverflow.com/questions/14507794/python-pandas-how-to-flatten-a-hierarchical-index-in-columns
    '''    
    df.columns = [
        '_'.join(tuple(map(str, t))).rstrip('_') 
        for t in df.columns.values
        ]
    return df

flatten_cols(df_pop)
df_pop = df_pop.reset_index()
df_pop.head()
```

```python
# Différence de population entre une année antérieure et l'année courante, exprimée en pourcentage
df_pop['var_pop_%'] = (df_pop['Population totale_2013'] - df_pop['Population totale_2012']) / df_pop['Population totale_2012'] * 100
df_pop.head()
```

## Classification hiérarchique des pays

Dans le domaine de la science des données, la classification consiste à créer des classes qui sont des groupes d'individus ayant des caractéristiques communes. Dans notre étude, nous sommes dans le cas d'une classification non-supervisée, c-à-d. que nous disposons d'éléments qui ne sont pas classés et on veut les regrouper entre eux.

Il y a deux grandes techniques de classification :
- le partitionnement
- la classification hiérarchique

Nous allons utiliser la classification hiérarchique bien qu'elle soit coûteuse en termes de mémoire et de calculs. Elle a l'avantage de nous laisser déterminer le nombre pertinent de "clusters" à-postériori. Par ailleurs, nous n'avons que peu de dimensions (une dizaine) et d'observations (moins de 200), l'algorithme sera par conséquent efficient.

Pour visualiser la classification hiérarchique, nous allons utiliser un dendrogramme. C'est un diagramme servant à représenter l'arrangement hiérarchique de groupes en se basant sur le degré de similarité que l'on détermine en introduisant des mesures de distance entre les objets. C'est la méthode de Ward qui sera préférée pour calculer les distances dans ce projet.

### Centrage-réduction et lancement de la classification hiérarchique

Pour le classification hiérarchique, je vais centrer-réduire les données car mes variables sont exprimées dans des unités différentes. Selon *Statsoft*, la standardisation (action de centrer-réduire) :
> fait référence à la transformation de données en soustrayant à chaque valeur une valeur de référence (classiquement une moyenne d'échantillon) et en la divisant par l'écart-type (typiquement un EcT d'échantillon). Cette importante transformation rendra toutes les valeurs (indifféremment de leurs distributions et unités de mesures originales) en unités compatibles avec distribution de moyenne 0 et d'écart-type 1.


Sources : 
- https://fr.wikipedia.org/wiki/Variable_centr%C3%A9e_r%C3%A9duit
- http://www.statsoft.fr/concepts-statistiques/glossaire/c/centrer.html

```python
# Suppression des attributs inutiles
df_hc = df_hc.drop(["pib_growth_(M$)", "fbcf_(M$)"], 1)

X = df_hc.values[:,2:] # On supprime les deux premières colonnes
print("X : ")
print(X)
print(X.shape)

# Centrage-réduction
X = X.astype(np.float64) # conversion des nombres entiers en nombres décimaux (float) : necessaire pour le StandardScaler
std_scaler = preprocessing.StandardScaler().fit(X) # Calcule la moyenne et l'écart-type pour le centrage-réduction
X_scaled = std_scaler.transform(X) # Centre et réduit
print("\nX_scaled : ")
print(X_scaled)
print(X_scaled.shape)

# Clustering hiérarchique
Z = linkage(X_scaled, 'ward')
print("\nZ : ")
print(Z)
print(Z.shape)
```

### Affichage et coupage du dendrogramme

```python
# Affichage du dendrogramme
plt.figure(figsize=(16, 9))
plt.ylabel('distance')
dendrogram(Z, no_labels=True)
plt.show()

# Coupage du dendrogramme en 5 clusters
final_clusters = fcluster(Z, t=5, criterion='maxclust')
df_hc['groupe_ch'] = final_clusters
print("Nb d'observations dans chaque groupe : ")
df_hc['groupe_ch'].value_counts()
```

## ACP et visualisation des partitions

L’analyse en composantes principales (ACP) sert à réduire le nombre de dimensions d'un jeu de données tout en gardant le plus d'informations. Elle transforme un certain nombre de variables (éventuellement) corrélées en un nombre (plus petit) de variables non-corrélées appelées composantes principales. La première composante principale explique autant que possible la variabilité dans les données, et chaque composante suivante explique autant que possible la variabilité restante.

Dans notre cas, nous effectuons une ACP pour visualiser les pays dans le premier plan factoriel. Ainsi, nous espérons mieux intérpréter nos partitions.

Sources :
- https://plot.ly/ipython-notebooks/principal-component-analysis/
- https://towardsdatascience.com/pca-using-python-scikit-learn-e653f8989e60
- https://www.youtube.com/watch?v=uV5hmpzmWsU
- https://www.youtube.com/watch?v=KrNbyM925wI&list=PLnZgp6epRBbRn3FeMdaQgVsFh9Kl0fjqX
- https://medium.com/data-design/how-to-not-be-dumb-at-applying-principal-component-analysis-pca-6c14de5b3c9d

```python
# Calcul des composantes principales
pca = decomposition.PCA()
pca.fit(X_scaled) # Ajuste le modèle avec X
X_projected = pca.transform(X_scaled) # Applique la réduction de dimensionnalité à X
print("\nX_projected : ")
print(X_projected)
print(X_projected.shape)

# Affichage des éboulis
eboulis = pca.explained_variance_ratio_*100
plt.bar(np.arange(len(eboulis))+1, eboulis)
plt.plot(np.arange(len(eboulis))+1, eboulis.cumsum(),c="red",marker='o')
plt.ylim(0,100)
plt.xlabel("rang de l'axe d'inertie")
plt.ylabel("pourcentage d'inertie")
plt.title("Eboulis des valeurs propres")
plt.show()
```

```python
# Coordonnées des individus
df_acp = df_hc[['pays', 'groupe_ch']].copy()
for i in range(0, 3) :
    df_acp['dim_' + str(i + 1)] = X_projected[:, i]

# Vérification de l'opération 
print(df_acp.head())

# Je suis obligé de créer une nouvelle colonne pour mon graphique
# Seaborn produit une légende fausse si j'use de 'groupe_ch'
# Plus de détails: https://github.com/mwaskom/seaborn/issues/1515
df_acp['num_groupe'] = df_acp['groupe_ch']
df_acp = df_acp.replace({
    'num_groupe' : {  
        1 : 'Un', 
        2 : 'Deux',    
        3 : 'Trois',
        4 : 'Quatre',
        5 : 'Cinq'
    }
})

# Représentation des individus sur le premier plan factoriel
f, ax = plt.subplots(figsize=(11, 9))
ax = sns.scatterplot(x="dim_1", y="dim_2",
                     hue="num_groupe", data=df_acp)
plt.title("Représentation des pays sur le premier plan factoriel")
plt.show()
```

```python
# Création du cercle des corrélations
pcs = pca.components_

# Définition de la taille du graphique
plt.figure(figsize=(14, 14))
corr_graph_size = [-0.48, 0.48]

for i, (x, y) in enumerate(zip(pcs[0, :], pcs[1, :])):
    # Afficher un segment de l'origine au point (x, y)
    plt.plot([0, x], [0, y], color='k')
    # Afficher le nom (data.columns[i]) de la performance
    plt.text(x, y, df_hc.columns[i + 2], fontsize='10')

    
# Afficher une ligne horizontale y=0
plt.plot(corr_graph_size, [0, 0], color='grey', ls='--')

# Afficher une ligne verticale x=0
plt.plot([0, 0], corr_graph_size, color='grey', ls='--')

plt.xlim(corr_graph_size)
plt.ylim(corr_graph_size)
plt.xlabel("F1 (" + str(round(pca.explained_variance_ratio_[0] * 100, 2)) + "%)")
plt.ylabel("F2 (" + str(round(pca.explained_variance_ratio_[1] * 100, 2)) + "%)")
plt.title("Cercle des corrélations")

plt.show()
```

## Test d'adéquation

Nous allons utiliser test de D'Agostino comme test d'adéquation. Il sert à vérifier si la distribution d'un échantillion ressemble à la distribution normale. L’hypothèse nulle est que les valeurs sont distribuées normalement.

Sources :
- https://machinelearningmastery.com/a-gentle-introduction-to-normality-tests-in-python/
- https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.normaltest.html

```python
from scipy import stats

k2, p_value = stats.normaltest(df_hc["dispo_prot/pers/j"])
alpha = 1e-3

print("p = {:g}".format(p_value))

if p_value < alpha: 
    print("L'hypothèse nulle peut être rejetée : la distribution ne suit pas la loi normale")
else:
    print("L'hypothèse nulle ne peut pas être rejetée : la distribution suit la loi normale")
    
df_hc[["dispo_prot/pers/j"]].hist()
plt.title("Distribution de la disponibilité de protéines en quantité (g/personne/jour)")
plt.show()
```

## Test de comparaison de deux populations (dans le cas gaussien)

Je vais utiliser deux tests de comparaisons :
- Je dois d'abord tester l'égalité des variances grâce au test de Bartlett. Il est nécessaire pour le test suivant.
- Ensuite je vais comparer deux moyennes dans le cas Gaussien grâce au test de Student.

Sources :
- https://docs.scipy.org/doc/scipy-0.15.1/reference/generated/scipy.stats.ttest_ind.html
- https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.bartlett.html#scipy.stats.bartlett
- https://stackoverflow.com/questions/25064506/what-scipy-statistical-test-do-i-use-to-compare-sample-means

```python
sample1 = df_hc.loc[df_hc["groupe_ch"] == 1,"dispo_prot/pers/j"]
sample2 = df_hc.loc[df_hc["groupe_ch"] == 3,"dispo_prot/pers/j"]

# Test d'égalité des variances
# https://en.wikipedia.org/wiki/Bartlett%27s_test

t_stat, p_value = stats.bartlett(sample1, sample2)
alpha = 1e-3

print("p = {:g}".format(p_value))

if p_value < alpha:  # null hypothesis: all input samples are from populations with equal variances
    print("L'hypothèse nulle peut être rejetée : les échantillions ont des variances significativement différentes")
else:
    print("L'hypothèse nulle ne peut pas être rejetée : les échantillions n'ont pas des variances significativement différentes")
    

# Affichage des deux distributions
plt.hist(sample1, alpha=0.5, label='1')
plt.hist(sample2, alpha=0.5, label='3')
plt.legend(loc='upper right')
plt.show()
```

```python
t_stat, p_value = stats.ttest_ind(sample1, sample2, equal_var=True)
alpha = 1e-3

print("p = {:g}".format(p_value))

if p_value < alpha: # null hypothesis: all input samples are from populations with equal variances
    print("L'hypothèse nulle peut être rejetée : les moyennes des deux populations sont significativement différentes")
else:
    print("L'hypothèse nulle ne peut pas être rejetée : les moyennes des populations ne sont pas significativement différentes")
```
